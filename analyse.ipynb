```python
#analyse des fichiers


import pandas as pd
import os

# Dossier contenant les fichiers CSV
directory = 'data/'

# Liste des années à vérifier
years = list(range(2005, 2023))

# Liste des types de fichiers à lire
file_types = ['caracteristiques', 'lieux', 'usagers', 'vehicules', 'vehicules-immatricules-baac']

# Dictionnaire pour stocker les informations sur les colonnes de chaque fichier
files_info = {file_type: {} for file_type in file_types}

for file_type in file_types:
    for year in years:
        file_path = os.path.join(directory, f'{file_type}_{year}.csv')
        if not os.path.exists(file_path):
            files_info[file_type][year] = {
                'file_name': f'{file_type}_{year}.csv',
                'num_columns': 0,
                'columns': [],
                'status': 'File not found'
            }
            continue

        separators = [',', ';', '\t']
        for sep in separators:
            try:
                df = pd.read_csv(file_path, encoding='ISO-8859-1', delimiter=sep, on_bad_lines='skip')
                if len(df.columns) > 1:
                    columns = df.columns.tolist()
                    files_info[file_type][year] = {
                        'file_name': f'{file_type}_{year}.csv',
                        'num_columns': len(columns),
                        'columns': columns,
                        'status': 'Success'
                    }
                    break  # Si nous avons trouvé un séparateur qui fonctionne, nous arrêtons les tentatives
            except Exception as e:
                continue  # Si une tentative échoue, nous essayons le séparateur suivant

        if year not in files_info[file_type]:
            files_info[file_type][year] = {
                'file_name': f'{file_type}_{year}.csv',
                'num_columns': 1,
                'columns': [],
                'status': 'Failed to read'
            }

# Affichage des informations sur les colonnes pour chaque fichier
for file_type, data in files_info.items():
    for year, info in data.items():
        print(f"Fichier {file_type}_{year}:")
        print(f"Nom du fichier: {info['file_name']}")
        print(f"Statut: {info['status']}")
        print(f"Nombre de colonnes: {info['num_columns']}")
        print(f"Noms des colonnes: {info['columns']}")
        print()

# Comparaison des colonnes pour chaque type de fichier
for file_type in file_types:
    print(f"Comparaison des colonnes pour le type de fichier: {file_type}")
    common_columns = None
    for year, info in files_info[file_type].items():
        if info['num_columns'] > 1:
            if common_columns is None:
                common_columns = set(info['columns'])
            else:
                common_columns = common_columns.intersection(info['columns'])
    
    if common_columns:
        print(f"Colonnes communes pour {file_type}: {common_columns}")
    else:
        print(f"Aucune colonne commune trouvée pour {file_type}.")

    uncommon_columns = {}
    for year, info in files_info[file_type].items():
        if info['num_columns'] > 1:
            uncommon_columns[year] = list(set(info['columns']) - common_columns)
    
    print("Colonnes qui ne sont pas communes :")
    for year, cols in uncommon_columns.items():
        if cols:
            print(f"Année {year}: {cols}")
    print()
```

V2 avec création d'un excel

```python
import pandas as pd
import os

# Dossier contenant les fichiers CSV
directory = 'data/'

# Liste des années à vérifier
years = list(range(2005, 2023))

# Liste des types de fichiers à lire
file_types = ['caracteristiques', 'lieux', 'usagers', 'vehicules', 'vehicules-immatricules-baac']

# Dictionnaire pour stocker les informations sur les colonnes et les lignes de chaque fichier
files_info = {file_type: {} for file_type in file_types}

for file_type in file_types:
    for year in years:
        file_path = os.path.join(directory, f'{file_type}_{year}.csv')
        if not os.path.exists(file_path):
            files_info[file_type][year] = {
                'file_name': f'{file_type}_{year}.csv',
                'num_columns': 0,
                'num_rows': 0,
                'columns': [],
                'status': 'File not found'
            }
            continue

        separators = [',', ';', '\t']
        for sep in separators:
            try:
                df = pd.read_csv(file_path, encoding='ISO-8859-1', delimiter=sep, on_bad_lines='skip')
                if len(df.columns) > 1:
                    columns = df.columns.tolist()
                    num_rows = len(df)
                    files_info[file_type][year] = {
                        'file_name': f'{file_type}_{year}.csv',
                        'num_columns': len(columns),
                        'num_rows': num_rows,
                        'columns': columns,
                        'status': 'Success'
                    }
                    break  # Si nous avons trouvé un séparateur qui fonctionne, nous arrêtons les tentatives
            except Exception as e:
                continue  # Si une tentative échoue, nous essayons le séparateur suivant

        if year not in files_info[file_type]:
            files_info[file_type][year] = {
                'file_name': f'{file_type}_{year}.csv',
                'num_columns': 1,
                'num_rows': 0,
                'columns': [],
                'status': 'Failed to read'
            }

# Préparation des données pour l'export en Excel
analysis_results = []
common_and_uncommon_info = []
total_lines_per_file_type = {file_type: 0 for file_type in file_types}

for file_type, data in files_info.items():
    common_columns = None
    for year, info in data.items():
        if info['num_columns'] > 1:
            if common_columns is None:
                common_columns = set(info['columns'])
            else:
                common_columns = common_columns.intersection(info['columns'])
            total_lines_per_file_type[file_type] += info['num_rows']

    uncommon_columns = {}
    for year, info in data.items():
        if info['num_columns'] > 1:
            uncommon_columns[year] = list(set(info['columns']) - common_columns)

    for year, info in data.items():
        analysis_results.append({
            'file_type': file_type,
            'year': year,
            'file_name': info['file_name'],
            'status': info['status'],
            'num_columns': info['num_columns'],
            'num_rows': info['num_rows'],
            'columns': ', '.join(info['columns']),
            'common_columns': ', '.join(common_columns) if common_columns else 'None',
            'uncommon_columns': ', '.join(uncommon_columns[year]) if year in uncommon_columns else 'None'
        })

    common_and_uncommon_info.append({
        'file_type': file_type,
        'common_columns': ', '.join(common_columns) if common_columns else 'None',
        'uncommon_columns': {year: ', '.join(cols) for year, cols in uncommon_columns.items() if cols}
    })

# Ajout de l'information sur le nombre total de lignes par type de fichier
total_lines_info = [{'file_type': file_type, 'total_lines': total} for file_type, total in total_lines_per_file_type.items()]

# Création d'un DataFrame pour les résultats d'analyse
analysis_df = pd.DataFrame(analysis_results)

# Création d'un DataFrame pour les informations communes et non communes
common_and_uncommon_df = pd.DataFrame(common_and_uncommon_info)

# Création d'un DataFrame pour les informations sur le nombre total de lignes
total_lines_df = pd.DataFrame(total_lines_info)

# Export des résultats d'analyse en fichier Excel
with pd.ExcelWriter('analysis_results.xlsx') as writer:
    analysis_df.to_excel(writer, sheet_name='Analysis Results', index=False)
    common_and_uncommon_df.to_excel(writer, sheet_name='Common and Uncommon Columns', index=False)
    total_lines_df.to_excel(writer, sheet_name='Total Lines', index=False)

print("Export terminé.")
```

Concatenation des csv

```python
import pandas as pd
import os

# Dossier contenant les fichiers CSV
directory = 'data/'

# Liste des années à vérifier
years = list(range(2005, 2023))

# Liste des types de fichiers à lire
file_types = ['caracteristiques', 'lieux', 'usagers', 'vehicules', 'vehicules-immatricules-baac']

# Dictionnaire pour stocker les DataFrames concaténés par type de fichier
concatenated_dfs = {file_type: [] for file_type in file_types}

for file_type in file_types:
    for year in years:
        file_path = os.path.join(directory, f'{file_type}_{year}.csv')
        if os.path.exists(file_path):
            separators = [',', ';', '\t']
            for sep in separators:
                try:
                    df = pd.read_csv(file_path, encoding='ISO-8859-1', delimiter=sep, on_bad_lines='skip')
                    if 'Num_Acc' in df.columns:
                        df['Index'] = df['Num_Acc']
                    elif 'Accident_Id' in df.columns:
                        df['Index'] = df['Accident_Id']
                    else:
                        raise ValueError(f"Le fichier {file_path} ne contient ni 'Num_Acc' ni 'Accident_Id'")
                    
                    concatenated_dfs[file_type].append(df)
                    break  # Si nous avons trouvé un séparateur qui fonctionne, nous arrêtons les tentatives
                except Exception as e:
                    continue  # Si une tentative échoue, nous essayons le séparateur suivant

# Concaténer les DataFrames pour chaque type de fichier et exporter en CSV
for file_type, df_list in concatenated_dfs.items():
    if df_list:
        combined_df = pd.concat(df_list, ignore_index=True)
        combined_df.set_index('Index', inplace=True)
        combined_df.to_csv(f'{file_type}_concatenated.csv')

print("Concaténation et exportation terminées.")
```
